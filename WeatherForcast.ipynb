```python
import tkinter as tk
from tkinter import font
from tkinter import ttk
import requests
import json
from tkinter import Spinbox
from PIL import Image,ImageTk

HEIGHT=700
WIDTH=700
key=-1

root=tk.Tk()
root.title("Welcome to Weather Forcast")

#api.openweathermap.org/data/2.5/weather?q={city name}&appid={your api key}

#api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}

jlink="https://pkgstore.datahub.io/core/country-list/data_json/data/8c458f2d15d9f2119654b29ede6e45b8/data_json.json"
country_codes=requests.get(jlink)
codes=country_codes.json()


clist=[]
language={"Afrikaans":"af", "Albanian": "al", "Arabic": "ar", "Azerbaijani": "az","English": "en", "French": "fr", "Hindi": "hi",
          "Indonesian": "id", "Italian": "it", "Japanese": "ja", "Korean": "kr", "Portuguese": "pt", "PortuguÃªs Brasil": "pt_br",
          "Romanian": "ro","Russian": "ru", "Dutch":" nl", "Chinese Simplified": "zh_cn", "Chinese Traditional": "zh_tw"}
for key in language:
    clist.append(key)
country=tuple(clist)

def testfunction(entry):
    print("this is an entry "+str(entry))

def format_response(weather):
    try:
        name=weather["name"]
        desc=weather["weather"][0]["description"]
        temp=weather["main"]["temp"]
        humid=weather["main"]["humidity"]
        time_zone=weather["timezone"]
        country=weather["sys"]["country"]
        for code in codes:
            if code["Code"]==country:
                cname=code["Name"]
        data="City: %s \nConditions: %s \nTemperature(celcius): %s\nTime zone: %s\nCountry: %s" %(name,desc, temp,time_zone,cname)
    except:
        data="Please enter valid city name!"
    #print(data)
    return data

def get_weather():
    name=city.get()
    lat=latitude.get()
    long=longitude.get()
    bhasha=language[number_chosen.get()]
    weather_key="6e7519c108dd77e893217f8e05bf2e08"
    url="https://api.openweathermap.org/data/2.5/weather"
    print("inside get_weather")
    if name!= "":
        params={"appid":weather_key,"q":name,"units":"Metric","lang":bhasha}
    elif lat!="" and long!="":
        params={"appid":weather_key,"lat":lat,"lon":long,"units":"Metric","lang":bhasha}
    else:
        label["text"]="Invalid parameters!"
        return
    response=requests.get(url, params)
    weather=response.json()
    icon_name = weather['weather'][0]['icon']
    label["text"]=format_response(weather)
    print(weather)
    open_image(icon_name)
    
    
def open_image(icon):
    size = int(lowerframe.winfo_height()*0.25)
    img = ImageTk.PhotoImage(Image.open('./img/'+icon+'.png').resize((size, size)))
    weather_icon.delete("all")
    weather_icon.create_image(0,0, anchor='nw', image=img)
    weather_icon.image = img
    
def get_by_city():
    key=0
    city.config(state="normal")
    city.insert(0,"Enter city:")
    city.bind('<FocusIn>', city_click)
    latitude.delete(0,"end")
    latitude.insert(0,"")
    latitude.config(state="disabled")
    longitude.delete(0,"end")
    longitude.insert(0,"")
    longitude.config(state="disabled")
    
    print(key)
def get_by_coordinates():
    key=1
    city.delete(0,"end")
    city.insert(0,"")
    city.config(state="disabled")
    
    
    latitude.config(state="normal")
    latitude.insert(0,"Latitude:")
    latitude.bind('<FocusIn>', latitude_click)
    
    longitude.config(state="normal")
    longitude.insert(0,"Longitude:")
    longitude.bind('<FocusIn>', longitude_click)
    print(key)
    
canvas=tk.Canvas(root, height=HEIGHT, width=WIDTH, bg="grey")
canvas.grid(column=0,row=0)

backimage=tk.PhotoImage(file="background3.png")
backlabel=tk.Label(root, image=backimage)
backlabel.place(x=0.75,y=0.75,relwidth=1, relheight=1)

upper_frame=tk.Frame(root, bg="#80c1ff", bd=5)
upper_frame.place(relwidth=0.75, relheight=0.1, relx=0.5, rely=0.1, anchor="n")

radVar=tk.IntVar()
curRad=tk.Radiobutton(upper_frame, text="By City",font=("courier",15, "bold"),bg="white",command=get_by_city, variable=radVar, value=1, anchor="n")
curRad.place(relwidth=0.30,relheight=0.75,relx=0.01,rely=0.075)

radVar2=tk.IntVar()
curRad2=tk.Radiobutton(upper_frame, text="By Coordinates",font=("courier",15, "bold"),bg="white",command=get_by_coordinates, variable=radVar, value=2, anchor="n")
curRad2.place(relwidth=0.60,relheight=0.75,relx=0.4,rely=0.075)


frame=tk.Frame(root,bg="#80c1ff", bd=5)
frame.place(relwidth=0.75, relheight=0.15, relx=0.5, rely=0.25, anchor="n")



def city_click(event):
    if city.get() == 'Enter city:':
        city.delete(0,"end")
        city.insert(0, '')
    
def latitude_click(event):
    if latitude.get() == 'Latitude:':
        latitude.delete(0,"end")
        latitude.insert(0, '')
    
def longitude_click(event):
    if longitude.get() == 'Longitude:':
        longitude.delete(0,"end")
        longitude.insert(0, '')
        

city=tk.Entry(frame, font=("courier",15,"italic","bold"), state="disabled")
city.place(relwidth=0.65, relheight=0.4)


latitude=tk.Entry(frame, font=("courier",15,"italic","bold"), state="disabled")
latitude.place(relwidth=0.30, relheight=0.4,rely=0.5)


longitude=tk.Entry(frame, font=("courier",15,"italic","bold"), state="disabled")
longitude.place(relwidth=0.30, relheight=0.4, relx=0.35, rely=0.5)



button=tk.Button(frame, text="Get weather", bg="white", font=40, command=get_weather)
button.place(relx=0.7,rely=0, relwidth=0.3)

number=tk.StringVar()
number_chosen=ttk.Combobox(frame,width=12,textvariable=number,state="readonly")
number_chosen['values']=country
number_chosen.place(relx=0.7, rely=0.5, relwidth=0.3)
number_chosen.current(4)

lowerframe=tk.Frame(root, bg="#80c1ff", bd=5)
lowerframe.place(relx=0.5,rely=0.45,relwidth=0.75,relheight=0.6, anchor="n")

label=tk.Label(lowerframe, text="Weather info", bg="white", font=("Modern",20,"bold"), bd=14, anchor="nw", justify="left")
label.place( relwidth=1, relheight=1)


weather_icon = tk.Canvas(label, bg="white", bd=0, highlightthickness=0)
weather_icon.place(relx=.75, rely=0, relwidth=1, relheight=0.5)



root.iconbitmap("sunny_sun_cloud_time_time_2777.ico")
root.mainloop()
```
